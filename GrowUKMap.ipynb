```python
#Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Loading the dataset into a DataFrame
file_path = 'D:/UoD/ProgrammingLanguagesforDataEngineering/Python_Grow/New folder/Growlocations.csv'
df = pd.read_csv(file_path)
```

```python
# Filtering out locations outside the allowed values for latitude and longitude
valid_latitude_range = (50.681, 57.985)
valid_longitude_range = (-10.592, 1.6848)
```

```python
#Selecting only the necessary columns to our dataframe
df = df[['Latitude', 'Longitude']]
```

```python
# Swapping latitude and longitude columns
df = df.rename(columns={'Latitude': 'Longitude', 'Longitude': 'Latitude'})
```

```python
# Filtering out invalid data points
df = df[(df['Latitude'] >= valid_latitude_range[0]) & (df['Latitude'] <= valid_latitude_range[1])]
```

```python
df = df[(df['Longitude'] >= valid_longitude_range[0]) & (df['Longitude'] <= valid_longitude_range[1])]
```

```python
# Plotting the PNG map
map_image = mpimg.imread('D:/UoD/ProgrammingLanguagesforDataEngineering/Python_Grow/New folder/map7.png')
```

```python
# Plot the map with the bounding box
plt.figure(figsize=(10, 6))
plt.imshow(map_image, extent=[valid_longitude_range[0], valid_longitude_range[1], valid_latitude_range[0], valid_latitude_range[1]], aspect='auto')
# Plot the sensor locations on the map
plt.scatter(df['Longitude'], df['Latitude'], color='red', s=10)
plt.title('Sensor Locations on Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
